```python
def check_invalid_data(data):
    col = data.columns.tolist()
    for i in range(len(col)):
        if data[col[i]].map(type) == str:
            print("Data in column", col[i], "is valid")
        elif data[col[i]].map(type) == float:
#             if dtype_dic[col[i]] == dtype('float64'):
            print("Data in column", col[i], "is valid")
        else:
            print("no invalid data")
    
```

```python
def check_duplicated(data: pd.DataFrame, primary_key: str) -> pd.DataFrame:
    duplicated = data[primary_key].duplicated()
    duplicate_df = data[duplicated]
    if duplicated.any().sum() > 0:
        print("Duplicated entries detected")
        return duplicate_df
    else:
        print("No duplicate detected")
    
    
```
